```python
# Import Required Libraries 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

train = pd.read_csv("Training.csv")

test = pd.read_csv("Testing.csv")
```

```python
train.info()
```

```python
train.sample(5)
```

```python
train = train.dropna(axis = 1)
train.sample(5)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
train['prognosis'] = encoder.fit_transform(train['prognosis'])
train.prognosis.describe()
```

```python
train.head()
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
X = train.iloc[:,:-1]
y = train.iloc[:,-1]
X_train, X_test, y_train, y_test =train_test_split(X, y, test_size = 0.2, random_state = 24)

print(f"Train: {X_train.shape}, {y_train.shape}")
print(f"Test: {X_test.shape}, {y_test.shape}")
```

```python
train.describe()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

# Initialize the model
model = Sequential()

# Input Layer and First Hidden Layer
model.add(Dense(256, input_dim=132, activation='relu'))
model.add(Dropout(0.3))  # Optional dropout for regularization

# Second Hidden Layer
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.3))

# Third Hidden Layer
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.2))

# Output Layer
model.add(Dense(41, activation='softmax'))  # 40 output neurons for 40 diseases

# Compile the model
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Summary of the model
model.summary()
```

```python
model.fit(X_train, y_train, epochs = 3,
           batch_size = 1, verbose = 1)
 
# Predicting the Value
y_pred = model.predict(X_test)
print(y_pred)
```

```python
loss, accuracy = model.evaluate(X_test, y_test, verbose=1)
print(f"Test Accuracy: {accuracy * 100:.2f}%")
```

```python
from tensorflow.keras.utils import to_categorical

# Ensure y_train and y_test are integers before encoding
y_test = to_categorical(y_test, num_classes=41)
```

```python
import numpy as np
from sklearn.metrics import accuracy_score

# Convert predictions to class labels
y_pred_classes = np.argmax(y_pred, axis=1)
y_test_classes = np.argmax(y_test, axis=1)

# Calculate accuracy
accuracy = accuracy_score(y_test_classes, y_pred_classes)
print(f"Test Accuracy: {accuracy * 100:.2f}%")
```
